```python
def hammingDistance(p, q):
    cnt = 0
    for i in range(len(p)):
        if p[i] != q[i]:
            cnt += 1
    return cnt
```

```python
p, q = input(), input()
print(hammingDistance(p, q))
```

```python
def approx_pattern_match(t, p, d):
    ids = []
    l = len(p)
    
    for i in range(l-1, len(t)):
        if hammingDistance(t[i-l+1:i+1], p) <= d:
            ids.append(i-l+1)

    return ids
```

```python
p, t, d = input(), input(), int(input())
for id in approx_pattern_match(t, p, d):
    print(id, end=" ")
```

```python
import itertools

def most_freq(t, k, d):
    dt = {}
    
    for s in map(''.join, itertools.product('ACTG', repeat=k)):
        dt[s] = 0
    
    for i in range(k-1, len(t)):
        s = t[i-k+1:i+1]
         
        for kk in dt:
            if hammingDistance(kk, s) <= d:
                dt[kk] += 1
    
    mx = max(dt.values())
    ss = []
    for k in dt:
        if dt[k] == mx:
            ss.append(k)
    return ss
```

```python
t, k, d = input(), int(input()), int(input())
for p in most_freq(t, k, d):
    print(p, end=" ")
```

```python
import itertools

def reverse_complement(s):
    ss = ""
    for ch in s:
        if ch == 'T':
            ss += 'A'
        elif ch == 'A':
            ss += 'T'
        elif ch == 'G':
            ss += 'C'
        else:
            ss += 'G'
    return ss[::-1]

def most_freq_plus(t, k, d):
    dt = {}
    
    for s in map(''.join, itertools.product('ACTG', repeat=k)):
        dt[s] = 0
    
    for i in range(k-1, len(t)):
        s = t[i-k+1:i+1]
         
        for kk in dt:
            if hammingDistance(kk, s) <= d:
                dt[kk] += 1
    
    mx = 0
    ss = []
    for k in dt:
        rev = reverse_complement(k)
        if dt[k]+dt[rev] > mx:
            mx = dt[k]+dt[rev]
            ss = [k]
        elif dt[k]+dt[rev] == mx:
            ss.append(k)
    print(mx)
    return ss
```

```python
t, k, d = input(), int(input()), int(input())
for p in most_freq_plus(t, k, d):
    print(p, end=" ")
```
