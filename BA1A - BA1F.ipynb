```python
def pattern_count(text: str, pattern: str) -> int:
    window = ""
    count = 0
    k = len(pattern)
    
    for index, ch in enumerate(text):
        if index >= k:
            window = window[1:]
        window += ch

        if window == pattern:
            count += 1
            
    return count
```

```python
text = input()
pattern = input()
print(pattern_count(text, pattern))
```

```python
def most_frequent_k_mer(text: str, k: int) -> list:
    pattern = ""
    most_frequent = []
    current_max = 0
    
    for index, ch in enumerate(text):
        pattern += ch
        if index >= k:
            pattern = pattern[1:]
        else:
            continue
        
        pat_count = pattern_count(text, pattern)
        if pat_count > current_max:
            current_max = pat_count
            most_frequent = [pattern]
        elif pat_count == current_max and pattern not in most_frequent:
            most_frequent.append(pattern)
        
    return most_frequent
```

```python
text = input()
k = int(input())
k_mers = most_frequent_k_mer(text, k)

for index, k_mer in enumerate(k_mers):
    if index:
        print(" ", end="")
    print(k_mer, end="")
    
print("")
```

```python
def reverse(ch: chr) -> chr:
    if ch == 'T':
        return 'A'
    elif ch == 'A':
        return 'T'
    elif ch == 'C':
        return 'G'
    return 'C'

def reverse_complement(pattern: str) -> str:
    rev_pattern = ""
    for ch in pattern:
        rev_pattern += reverse(ch)
    return rev_pattern[::-1]
```

```python
text = input()
print(reverse_complement(text))
```

```python
def pattern_occurrences(text: str, pattern: str) -> list:
    k = len(pattern)
    occurrences = []
    window = ""
    
    for index, ch in enumerate(text):
        window += ch
        if index >= k:
            window = window[1:]
            if window == pattern:
                occurrences.append(index-k+1)
                
    return occurrences
```

```python
pattern = input()
text = input()
occurrences = pattern_occurrences(text, pattern)

for index, id in enumerate(occurrences):
    if index:
        print(" ", end="")
    print(id, end="")
print("")
```

```python
def find_Lt_clumps(genome: str, L: int, t: int, k: int) -> list:
    window = ""
    clumps = set()
    map_kmer_idx = {}
    
    for index, ch in enumerate(genome):
        window += ch
        if index >= k:
            window = window[1:]
                        
            if window not in map_kmer_idx:
                map_kmer_idx[window] = []
            
            r = len(map_kmer_idx[window])
            
            if r >= t:
                map_kmer_idx[window].pop(0)
            map_kmer_idx[window].append(index)

            if r == t-1 and map_kmer_idx[window][r] - map_kmer_idx[window][0] + 1 <= L:
                clumps.add(window)

    return clumps
```

```python
genome = input()
k, L, t = [int(x) for x in input().split()]

for k_mer in find_Lt_clumps(genome, L, t, k):
    print(k_mer, end=" ")
```

```python
def skew(genome: str) -> list:
    mn = 0
    min_list = []
    cur = 0
    
    for index, ch in enumerate(genome):
        if ch == 'C':
            cur -= 1
        elif ch == 'G':
            cur += 1
        
        if cur < mn:
            mn = cur
            min_list = [index+1]
        elif cur == mn:
            min_list.append(index+1)
            
    return min_list
```

```python
genome = input()
min_list = skew(genome)

for idx in min_list:
    print(idx, end=" ")
```
